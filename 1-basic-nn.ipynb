```python
import torch
import torch.nn as nn
```

```python
model = nn.Sequential(
    nn.Linear(4, 8),
    nn.Linear(8,5),
    nn.Linear(5, 3),
    nn.Linear(3,2),
    nn.Sigmoid()
)
```

```python
input = torch.Tensor([[4,9,2,7]])
```

```python
output = model(input)
print(output)
```

```python
target = torch.tensor([0], dtype=torch.long)
```

```python
criterion = nn.CrossEntropyLoss()
```

```python
loss = criterion(output, target)
print(loss)
```

```python
loss.backward()
```

```python
weight = model[0].weight
print(weight)
```

```python
weight_grad = weight.grad
print(weight_grad)
```

```python
bias = model[0].bias
print(bias)
```

```python
bias_grad = bias.grad
print(bias_grad)
```

```python
learning_rate = 0.01
```

```python
with torch.no_grad():
    model[0].weight.copy_(model[0].weight - learning_rate * model[0].weight.grad)
    model[0].bias.copy_(model[0].bias - learning_rate * model[0].bias.grad)

print("Model parameters updated in-place.")
```

```python
output = model(input)
print("New Output:", output)
loss = criterion(output, target)
print("New Loss:", loss)
```
